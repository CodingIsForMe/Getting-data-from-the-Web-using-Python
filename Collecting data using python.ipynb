```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get("https://www.geeksforgeeks.org/")
```

```python
soup = BeautifulSoup(req.content, "html.parser")

print(soup.prettify())
```

```python
print(soup.get_text())
```

```python
res = soup.title

print(res.prettify())
```

```python
print(res.get_text())
```

```python
rest = soup.body.div
print(rest.get_text())
```
